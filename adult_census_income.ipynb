```python
features = data.iloc[:,:-1]
label = data['income']
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
label = enc.fit_transform(label)
```

```python
features['age'][features['age'] < 18] = 0
features['age'][(features['age'] >= 18) & (features['age'] < 30)] = 1
features['age'][(features['age'] >= 30) & (features['age'] < 40)] = 2
features['age'][(features['age'] >= 40) & (features['age'] < 50)] = 3
features['age'][(features['age'] >= 50) & (features['age'] < 60)] = 4
features['age'][features['age'] >= 60] = 5
```

```python
features['fnlwgt'][(features['fnlwgt'] >= 12285.) & (features['fnlwgt'] < 159527.)] = 0
features['fnlwgt'][(features['fnlwgt'] >= 159527.) & (features['fnlwgt'] < 306769.)] = 1
features['fnlwgt'][(features['fnlwgt'] >= 306769.) & (features['fnlwgt'] < 454011.)] = 2
features['fnlwgt'][features['fnlwgt'] >= 454011.] = 3
```

```python
features['capital.gain'][features['capital.gain'] < 9999.9] = 0
features['capital.gain'][(features['capital.gain'] >= 9999.9) & (features['capital.gain'] < 19999.8)] = 1
features['capital.gain'][(features['capital.gain'] >= 19999.8) & (features['capital.gain'] < 29999.7)] = 2
features['capital.gain'][features['capital.gain'] > 29999.7] = 3
```

```python
features['capital.loss'][features['capital.loss'] < 435.6] = 0
features['capital.loss'][features['capital.loss'] >= 435.6] = 1
```

```python
features['hours.per.week'][features['hours.per.week'] < 10.8] = 0
features['hours.per.week'][(features['hours.per.week'] >= 10.8) & (features['hours.per.week'] < 40.2)] = 1
features['hours.per.week'][(features['hours.per.week'] >= 40.2) & (features['hours.per.week'] < 69.6)] = 2
features['hours.per.week'][features['hours.per.week'] >= 69.6] = 3
```

```python
dummies_workclass = pd.get_dummies(features['workclass'], prefix='workclass')
```

```python
dummies_education = pd.get_dummies(features['education'], prefix='education')
dummies_marital_status = pd.get_dummies(features['marital.status'], prefix='marital.status')
dummies_occupation = pd.get_dummies(features['occupation'], prefix='occupation')
dummies_relationship = pd.get_dummies(features['relationship'], prefix='relationship')
dummies_race = pd.get_dummies(features['race'], prefix='race')
dummies_sex = pd.get_dummies(features['sex'], prefix='sex')
```

```python
features['native.country'][features['native.country'] == 'United-States'] = 0
features['native.country'][features['native.country'] != 'United-States'] = 1
```

```python
features.head()
```

```python
features = features.drop(['workclass','education','marital.status','occupation','relationship','race','sex'], axis=1)
```

```python
features.head()
```

```python
features = features.join([dummies_workclass, dummies_education, dummies_marital_status, dummies_occupation, dummies_relationship, dummies_race, dummies_sex])
```

```python
features.head()
```

```python
# ROC
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import roc_auc_score, roc_curve, accuracy_score, auc
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.3, random_state=0)
```

```python
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_predict = clf.predict(X_test)
```

```python
roc_auc_score(y_test, clf.predict_proba(X_test)[:,1])
```

```python
accuracy_score(y_test, y_predict)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, clf.predict_proba(X_test)[:,1])
```

```python
auc(fpr, tpr)
```

```python
models = {
    'LogisticRegression':LogisticRegression(),
    'SVC':SVC(kernel='linear'),
    'DecisionTreeClassifier':DecisionTreeClassifier(max_depth=4),
    'RandomForestClassifier':RandomForestClassifier(max_depth=4, n_estimators=50, random_state=0),
    'AdaBoostClassifier':AdaBoostClassifier(n_estimators=50, random_state=0)
}
```

```python
train_model_eval(models, X_train, X_test, y_train, y_test)
```
